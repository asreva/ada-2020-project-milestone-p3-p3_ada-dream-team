**Study paper**: Eunjoon Cho, Seth A. Myers and Jure Leskovec, *Friendship and Mobility: User Movement in Location-Based Social Network*, Standford University, 2011

**Due date**: 6 November 2020, EOD

**Method**: The method that will be used here is the following:

0. Apply some preprocessing to clean data and remove useless information;
1. Convert long-lat check-ins' positions to x-y position (Mercator);
2. Assign to each check-in its square in the map;
3. Find for each user the square that has more check-ins;
4. Mean all the check-ins in that square to estimate home position;
5. Convert homes position from x-y to long-lat;
6. Compute the distance (Haversine) between two friends homes;
7. Estimate the probability distribution of the distance with a Gaussian Kernel Density Estimator;
8. Plot the result.

## 0. Extraction of the data and preprocessing

```python
#Librairies
import pandas as pd

#Constants
PATH = "data/"
EXTENSION = ".txt.gz"
SEPARATOR = "\t"
CHECKIN_COLUMNS_NAME = ["user", "check-in time", "latitude", "longitude", "location id"]
EDGE_COLUMNS_NAME = ["friend1", "friend2"]
```

```python
#Extract dataframe from .txt.gz (tabulated separated without header)
loc_brightkite_edges = pd.read_csv(PATH+"loc-brightkite_edges"+EXTENSION, sep=SEPARATOR, header = None, names = EDGE_COLUMNS_NAME)
loc_brightkite_totalCheckins = pd.read_csv(PATH+"loc-brightkite_totalCheckins"+EXTENSION, sep=SEPARATOR, header = None, names = CHECKIN_COLUMNS_NAME)
loc_gowalla_edges = pd.read_csv(PATH+"loc-gowalla_edges"+EXTENSION, sep=SEPARATOR, header = None, names = EDGE_COLUMNS_NAME)
loc_gowalla_totalCheckins = pd.read_csv(PATH+"loc-gowalla_totalCheckins"+EXTENSION, sep=SEPARATOR, header = None, names = CHECKIN_COLUMNS_NAME)
```

```python
#Sample dataframe to analyze and verify structure
#loc_brightkite_edges.sample(10)
loc_brightkite_totalCheckins.sample(10)
#loc_gowalla_edges.sample(10)
#loc_gowalla_totalCheckins.sample(10)
```

```python
#Remove from checkin the useless information (location ID and check-in time)
TO_REMOVE = ["check-in time", "location id"]
bright_checkins = loc_brightkite_totalCheckins.drop(columns=TO_REMOVE)
gowalla_checkins = loc_gowalla_totalCheckins.drop(columns=TO_REMOVE)

#Remove from checkin some invalid points (null data and incoherent latitudes)
bright_0_lat = bright_checkins[bright_checkins["latitude"] == 0]
bright_0_lat_long = bright_0_lat[bright_0_lat["longitude"] == 0]
bright_checkins = bright_checkins.drop(bright_0_lat_long.index)
bright_big_lat = bright_checkins[bright_checkins["latitude"]>90]
bright_checkins = bright_checkins.drop(bright_big_lat.index)
bright_small_lat = bright_checkins[bright_checkins["latitude"]<-90]
bright_checkins = bright_checkins.drop(bright_small_lat.index)

gowalla_0_lat = gowalla_checkins[gowalla_checkins["latitude"] == 0]
gowalla_0_lat_long = gowalla_0_lat[gowalla_0_lat["longitude"] == 0]
gowalla_checkins = gowalla_checkins.drop(gowalla_0_lat_long.index)
gowalla_big_lat = gowalla_checkins[gowalla_checkins["latitude"]>90]
gowalla_checkins = gowalla_checkins.drop(gowalla_big_lat.index)
gowalla_small_lat = gowalla_checkins[gowalla_checkins["latitude"]<-90]
gowalla_checkins = gowalla_checkins.drop(bright_small_lat.index)

#Rename friendships data
bright_edges = loc_brightkite_edges
gowalla_edges = loc_gowalla_edges
```

## 1. Compute home position

### a) Determine the 25km square with most check-ins
Data are projected to a Mercator plan, where squares are easy to define. The square with most check-ins is extracted.

```python
#Librairies
from pyproj import Transformer
import numpy as np

#Constants
SQUARE_SIZE = 25000

#Define the transformation from long-lat to x-y
trans = Transformer.from_crs(
    "epsg:4326", #long lat notation
    "epsg:3857", #pseudo mercator https://epsg.io/3857
    always_xy=True,
)

#Transform data long-lat -> x-y
bright_checkins["x"], bright_checkins["y"] = trans.transform(bright_checkins["longitude"].values, 
                                                             bright_checkins["latitude"].values)
gowalla_checkins["x"], gowalla_checkins["y"] = trans.transform(gowalla_checkins["longitude"].values, 
                                                             gowalla_checkins["latitude"].values)

#Some checkins at extreme values cannot be converted due to the projection -> remove them
bright_checkins = bright_checkins.replace([np.inf, -np.inf], np.nan).dropna(axis=0)
gowalla_checkins = gowalla_checkins.replace([np.inf, -np.inf], np.nan).dropna(axis=0)

#find square position to which the point belongs (take the divisor of the euclidian division) 
#(!i_x/i_y may be negative because the origin is at the center of Earth (Null Island))
bright_checkins["i_x"] = (bright_checkins["x"]/SQUARE_SIZE).astype(int)
bright_checkins["i_y"] = (bright_checkins["y"]/SQUARE_SIZE).astype(int)
gowalla_checkins["i_x"] = (gowalla_checkins["x"]/SQUARE_SIZE).astype(int)
gowalla_checkins["i_y"] = (gowalla_checkins["y"]/SQUARE_SIZE).astype(int)

#Now we can drop x, y, latitude and longitude
TO_REMOVE = ["latitude", "longitude","x"] #keep one column for the count
bright_checkins_light = bright_checkins.drop(columns=TO_REMOVE)
gowalla_checkins_light = gowalla_checkins.drop(columns=TO_REMOVE)
bright_checkins_light = bright_checkins_light.rename(columns={"y":"count"})
gowalla_checkins_light = gowalla_checkins_light.rename(columns={"y":"count"})
```

```python
#For each user find the most used square

#Group by user and square position and count the nb of occurence
bright_checkins_zone_count = bright_checkins_light.groupby(["user","i_x","i_y"]).count().reset_index()
#create a dataframe with the max value
bright_checkins_zone_max = bright_checkins_zone_count.groupby(["user"]).max() \
                           .rename(columns={"count":"count_max"}).drop(columns=["i_x", "i_y"])
#merge both dataframes
bright_checkins_merge = pd.merge(bright_checkins_zone_max, bright_checkins_zone_count, left_on="user",right_on="user")
#keep only the lines where the count is equal to the maximum
bright_checkins_home = bright_checkins_merge[(bright_checkins_merge["count_max"] == bright_checkins_merge["count"])]
#There may be still some duplicates if tie (keep first)
bright_checkins_home = bright_checkins_home.drop_duplicates(subset=["user"],keep="first")
#by rehandling the dataframe we have for each user it's prefered square
bright_square_home = bright_checkins_home.drop(columns=["count_max","count"]).set_index("user")

#Group by user and square position and count the nb of occurence
gowalla_checkins_zone_count = gowalla_checkins_light.groupby(["user","i_x","i_y"]).count().reset_index()
#create a dataframe with the max value
gowalla_checkins_zone_max = gowalla_checkins_zone_count.groupby(["user"]).max() \
                           .rename(columns={"count":"count_max"}).drop(columns=["i_x", "i_y"])
#merge both dataframes
gowalla_checkins_merge = pd.merge(gowalla_checkins_zone_max, gowalla_checkins_zone_count, left_on="user",right_on="user")
#keep only the nb of checkins of the square is equal to the maximum
gowalla_checkins_home = gowalla_checkins_merge[(gowalla_checkins_merge["count_max"] == gowalla_checkins_merge["count"])]
#There may be still some duplicates if tie (keep first)
gowalla_checkins_home = gowalla_checkins_home.drop_duplicates(subset=["user"],keep="first")
#by rehandling the dataframe we have for each user it's prefered square
gowalla_square_home = gowalla_checkins_home.drop(columns=["count_max","count"]).set_index("user")
```

### b) Determine the position of the home
Remove all check-ins that are not in most used square, mean the position of the remainings check-ins of each user and then reproject to the long-lat space (no loss in precision, the transformation is reversible + we could not have used mean consistently in long-lat space due to the nature of the unit) 

```python
#For each row add the home square of the user
bright_checkins_full = pd.merge(bright_checkins, bright_square_home, left_on="user", 
                                right_index=True, how="outer", suffixes=('', '_h'))
gowalla_checkins_full = pd.merge(gowalla_checkins, gowalla_square_home, left_on="user", 
                                right_index=True, how="outer", suffixes=('', '_h'))

#Remove checkins not in home square
bright_home_checkins = bright_checkins_full[(bright_checkins_full["i_x"] == bright_checkins_full["i_x_h"]) 
                                            & (bright_checkins_full["i_y"] == bright_checkins_full["i_y_h"])]
gowalla_home_checkins = gowalla_checkins_full[(gowalla_checkins_full["i_x"] == gowalla_checkins_full["i_x_h"]) 
                                            & (gowalla_checkins_full["i_y"] == gowalla_checkins_full["i_y_h"])]

#Remove useless columns
TO_REMOVE = ["i_x", "i_y", "i_x_h", "i_y_h", "longitude", "latitude"]
bright_home_checkins_light = bright_home_checkins.drop(columns=TO_REMOVE)
gowalla_home_checkins_light = gowalla_home_checkins.drop(columns=TO_REMOVE)

#Mean the position for each user
bright_home_pos = bright_home_checkins_light.groupby(["user"]).mean()
gowalla_home_pos = gowalla_home_checkins_light.groupby(["user"]).mean()

#Define the transformation from x-y to long-lat (return to long lat for more precise distance mesure) (linear trans -> ok)
trans = Transformer.from_crs(
    "epsg:3857", #pseudo mercator https://epsg.io/3857
    "epsg:4326", #long lat notation
    always_xy=True,
)

#Transform data from x-y to long-lat
long, lat = trans.transform(bright_home_pos["x"].values, bright_home_pos["y"].values)
bright_home_pos["lat"] = lat
bright_home_pos["long"] = long
long, lat = trans.transform(gowalla_home_pos["x"].values, gowalla_home_pos["y"].values)
gowalla_home_pos["lat"] = lat
gowalla_home_pos["long"] = long

#Now we can drop x and y
TO_REMOVE = ["x", "y"]
bright_home_pos_light = bright_home_pos.drop(columns=TO_REMOVE)
gowalla_home_pos_light = gowalla_home_pos.drop(columns=TO_REMOVE)
```

## 2. Compute distance between friends homes position
Add to the list of friends their home position and compute the Haversine distance (https://en.wikipedia.org/wiki/Haversine_formula) between their home. (There is no need to check if pairs of friends appear one or two times because this will not affect the probability distribution)

```python
#Library
import numpy as np

#Constants
R = 6371000 #radius of the earth, in meters

#Create a structure with the friend relationships and their homes position
bright_friend_homes = pd.merge(loc_brightkite_edges, bright_home_pos_light, left_on="friend1", right_index=True, how="outer")
bright_friend_homes = bright_friend_homes.rename(columns={"lat":"lat1","long":"long1"})
bright_friend_homes = pd.merge(bright_friend_homes, bright_home_pos_light, left_on="friend2", right_index=True, how="outer")
bright_friend_homes = bright_friend_homes.rename(columns={"lat":"lat2","long":"long2"})
gowalla_friend_homes = pd.merge(loc_gowalla_edges, gowalla_home_pos_light, left_on="friend1", right_index=True, how="outer")
gowalla_friend_homes = gowalla_friend_homes.rename(columns={"lat":"lat1","long":"long1"})
gowalla_friend_homes = pd.merge(gowalla_friend_homes, gowalla_home_pos_light, left_on="friend2", right_index=True, how="outer")
gowalla_friend_homes = gowalla_friend_homes.rename(columns={"lat":"lat2","long":"long2"})

#Because we removed checkout during the process their may exist user without home position so we have to remove them
bright_friend_homes = bright_friend_homes.dropna()
gowalla_friend_homes = gowalla_friend_homes.dropna()

#Remove useless columns
TO_REMOVE = ["friend1", "friend2"]
bright_friend_homes_light = bright_friend_homes.drop(columns=TO_REMOVE)
gowalla_friend_homes_light = gowalla_friend_homes.drop(columns=TO_REMOVE)

#Compute the distance using the Haversine distance
#Transform long/lat to radians, create delta values
lat1_r = np.radians(bright_friend_homes_light["lat1"])
lat2_r = np.radians(bright_friend_homes_light["lat2"])
lon1_r = np.radians(bright_friend_homes_light["long1"])
lon2_r = np.radians(bright_friend_homes_light["long2"])
d_lat = lat2_r - lat1_r
d_lon = lon2_r - lon1_r

#Formula
a = np.sin(d_lat / 2)**2 + np.cos(lat1_r) * np.cos(lat2_r) * np.sin(d_lon / 2)**2
c = 2 * np.arcsin(np.sqrt(a))
dist = R*c

#Convert to km
dist_km_bright = dist/1000

#Transform long/lat to radians, create delta values
lat1_r = np.radians(gowalla_friend_homes_light["lat1"])
lat2_r = np.radians(gowalla_friend_homes_light["lat2"])
lon1_r = np.radians(gowalla_friend_homes_light["long1"])
lon2_r = np.radians(gowalla_friend_homes_light["long2"])
d_lat = lat2_r - lat1_r
d_lon = lon2_r - lon1_r

#Formula
a = np.sin(d_lat / 2)**2 + np.cos(lat1_r) * np.cos(lat2_r) * np.sin(d_lon / 2)**2
c = 2 * np.arcsin(np.sqrt(a))
dist = R*c

#Convert to km
dist_km_gowalla = dist/1000
```

## 3. Plot probabilty distribution
The probability distribution is estimated using a gaussian kernel density.

```python
#Library
from sklearn.neighbors import KernelDensity
import matplotlib.pyplot as plt

#Constants
N_PROB_POINTS = 70
PAPER_FIG_ASPECT_RATION = 2.9/2.4

#Create and fit the kernel density
kd = KernelDensity(bandwidth=2, kernel='gaussian')
data = dist_km_bright.values.reshape((len(dist_km_bright.values), 1))
kd.fit(data)

#Compute probabilty for some values
dist_space = np.logspace(0, 5, N_PROB_POINTS)
dist_space = dist_space.reshape((len(dist_space), 1))
density_bright = np.exp(kd.score_samples(dist_space))

#Create and fit the kernel density
kd_2 = KernelDensity(bandwidth=2, kernel='gaussian')
data = dist_km_gowalla.values.reshape((len(dist_km_gowalla.values), 1))
kd_2.fit(data)

#Compute probabilty for some values
density_gowalla = np.exp(kd_2.score_samples(dist_space))

#Plot
plt.figure(figsize=(5*PAPER_FIG_ASPECT_RATION,5))
plt.title("Replication of Figure 2a) from Friendship and Mobility paper")
plt.plot(dist_space[:], density_bright, "o", fillstyle = "none", color = "blue", label = "Brightkite")
plt.plot(dist_space[:], density_gowalla, "*", fillstyle = "none", color = "red", label = "Gowalla")
plt.ylim((10**-6, 10**0))
plt.xlim((10**0, 10**5))
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Distance between homes (km)")
plt.ylabel("Probability")
plt.legend()
plt.show()
```

## 4. Conclusion

In this milestone, I replicated the figure 2a of *Friendship and Mobility*. To do so, I did the following:

0. Apply some preprocessing to clean data and remove useless information;
1. Convert long-lat check-ins' positions to x-y position (Mercator);
2. Assign to each check-in its square in the map;
3. Find for each user the square that has more check-ins;
4. Mean all the check-ins in that square to estimate home position;
5. Convert homes position from x-y to long-lat;
6. Compute the distance (Haversine) between two friends homes;
7. Estimate the probability distribution of the distance with a Gaussian Kernel Density Estimator;
8. Plot the result.

The plot obtained is quite similar to the one presented in the paper. Qualitatively the results are identical: the behaviour of the plot is the same, and we can notice the kink around 100km. Quantitatively the result is also really near to what the authors presented. However, there may be a little difference around the 1km values where the probability seems higher in their plot (around ~0.1 wherein my plot the probability is more about ~0.05). We cannot be sure that there are differences or not as it is hard to say due to the size of their plot, the fact that they don't display the 0.1 probabilty value and that there is no interpolation between the points. However, the difference may come from some difference in assumptions or some errors.

To replicate their data, some assumption had to be done due to a lack of information about the method that they used:
1. The preprocessing cleaning step may be different;
2. The square position assignment: I projected the data to Mercator in order to find the home square and then projected back its position;
3. During this projection the squares were created around a reference position that is the centre of long-lat system (Null Island);
4. When there was a tie between two possible square home we kept the first one in our order, but they may have done differently;
5. I computed the distance between two points with Haversine distance, they may have used an other metric,
6. A Gaussian Kernel Density was used as probability distribution estimator, they did not mentionned how they computed the distribution.

Some numerical error may have come from:
1. Conversion from long-lat to x-y and reverse (some extremes values had to be drop out and there exist imprecisions);
2. Computation of the distance between the points.

In terms of computation efficiency, on my computer, the code takes about a little more than 1 minute to run (1 minute 10 seconds). The most consuming blocks of code are the data loading (due to the size of the data to load) and the probabilty estimator (due to the complexity of the task with gaussian kernel and the size of the dataset). This duration seems reasonable to me knowing the size of the dataset.

To conclude, I was able to reproduce quite reliably their results, asserting the existence of the kink around 100km.
